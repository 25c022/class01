```python
# prompt: 움직이는 로봇 청소기인데 벽에 튕겨 그림으로 나오게 해줘.

import matplotlib.pyplot as plt
import random

# Parameters
width = 100  # Width of the room
height = 100 # Height of the room
robot_size = 5
num_steps = 200
speed = 5

# Initial position and direction
x, y = random.randint(0, width), random.randint(0, height)
dx, dy = random.uniform(-1, 1), random.uniform(-1, 1)

# Normalize direction vector
magnitude = (dx**2 + dy**2)**0.5
dx /= magnitude
dy /= magnitude

# Store trajectory
trajectory_x = [x]
trajectory_y = [y]

# Simulation
for _ in range(num_steps):
  # Update position
  x += dx * speed
  y += dy * speed

  # Bounce off walls
  if x <= 0 or x >= width:
    dx *= -1
  if y <= 0 or y >= height:
    dy *= -1

  # Store position
  trajectory_x.append(x)
  trajectory_y.append(y)

# Plotting
plt.figure(figsize=(8, 8))
plt.plot(trajectory_x, trajectory_y, marker='o', linestyle='-', markersize=robot_size/2, linewidth=0.5)
plt.xlim(0, width)
plt.ylim(0, height)
plt.title("Robot Vacuum Trajectory (Bouncing off Walls)")
plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
