```python
# prompt: 움직이는 로봇 청소기인데 정사각형을 그려.

import matplotlib.pyplot as plt

def draw_square(side_length):
    """Simulates a robot vacuum cleaner drawing a square and records positions."""
    positions = [(0, 0)]  # Start at the origin
    current_x, current_y = 0, 0

    # Move right
    for _ in range(side_length):
        current_x += 1
        positions.append((current_x, current_y))

    # Move down
    for _ in range(side_length):
        current_y -= 1
        positions.append((current_x, current_y))

    # Move left
    for _ in range(side_length):
        current_x -= 1
        positions.append((current_x, current_y))

    # Move up
    for _ in range(side_length):
        current_y += 1
        positions.append((current_x, current_y))

    return positions

# Simulate drawing a square with side length 5
side = 5
robot_path = draw_square(side)

# You can now use matplotlib to plot the path
x_coords, y_coords = zip(*robot_path)

plt.figure(figsize=(6, 6))
plt.plot(x_coords, y_coords, marker='o', linestyle='-')
plt.title(f'Robot Vacuum Drawing a Square of Side Length {side}')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box') # Ensure square aspect ratio
plt.show()
```
